```python
import folium
import geopandas

%matplotlib inline
```

```python
world = geopandas.read_file(
    geopandas.datasets.get_path('naturalearth_lowres')
)
    # NaturalEarth is a free dataset that comes with geopandas.
    #    See: http://www.naturalearthdata.com/ 
    # We could have just used a string with the filename of a shapefile there,
    #    instead of geopandas.datasets.get_path('naturalearth_lowres').

world.head(10)
```

```python
world['GDPpercapita'] = 10**6 * world['gdp_md_est']/world['pop_est']
# The 10**6 is just there because the units of gdp_md_est are 'millions of dollars' and 
#   I want the GDP/capita to be in $/person not millions-of-$/person.

world.head(4)
```

```python
world.plot(column="GDPpercapita",scheme='quantiles',k=9,legend=True,linewidth=0.0,figsize=(20,15),cmap="Reds")
```

```python
CH_Prov = geopandas.read_file("C:\\Users\\CAPITAL\\Desktop\\Maps\\China Shapefiles\\AdminandShape\\CHN_adm1.shp")

CH_Prov.head(34) #first seven entries
```

```python
import pandas
CHN_Table=pandas.read_csv("C:\\Users\\CAPITAL\\Dropbox\\Data\\China Economy\\ProvinceEnterpriseTotal2010-2016.csv")
CHN_Table.head(7)
```

```python
CH_Prov.rename(columns={'NAME_1':'Province'},inplace=True)
CHN_Table['E_TotChange']=CHN_Table['Enterprise_2016']-CHN_Table['Enterprise_2010']
```

```python
JOINING_KEY = 'Province'
CH_Prov_Enter = CH_Prov.merge(CHN_Table, on=JOINING_KEY)
CH_Prov_Enter.drop('CCN_1', axis=1, inplace=True)
CH_Prov_Enter.drop('CCA_1', axis=1, inplace=True)
CH_Prov_Enter.drop('TYPE_1', axis=1, inplace=True)
CH_Prov_Enter.drop('ENGTYPE_1', axis=1, inplace=True)
CH_Prov_Enter.drop('NL_NAME_1', axis=1, inplace=True)
CH_Prov_Enter.drop('VARNAME_1', axis=1, inplace=True)

CH_Prov_Enter.head(2)  
```

```python
CH_Prov_Enter.plot(column="E_PercentChange",scheme='quantiles',k=7,legend=True,linewidth=0.1,edgecolor="black",figsize=(20,15),cmap="RdBu")
```

```python
CH_Prov_Enter.crs
CH_Prov_Enter2=CH_Prov_Enter.to_crs(epsg=3857)
CH_Prov_Enter2.plot(column="E_PercentChange",scheme='quantiles',k=7,legend=True,linewidth=.1,edgecolor='black',figsize=(20,15),cmap="RdBu")
```
